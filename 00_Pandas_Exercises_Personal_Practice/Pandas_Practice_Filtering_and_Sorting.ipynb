# Ex1 - Filtering and Sorting Data
## This time we are going to pull data directly from the internet. Special thanks to: https://github.com/justmarkham for sharing the dataset and materials.

### Step 1. Import the necessary libraries

```python
import numpy as np
import pandas as pd
```

### Step 2. Import the dataset from this address: https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv
### Step 3. Assign it to a variable called chipo.

```python
url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv'
chipo = pd.read_table(url, sep='\t')
chipo.head()
```

#### clean the item_price column and transform it in a float

```python
chipo['item_price'] = chipo['item_price'].replace('[\$,]', '', regex=True).astype(float)
```

```python
chipo.head()
```

### delete the duplicates in item_name and quantity

```python
chipo_filtered = chipo.drop_duplicates(['quantity', 'item_name', 'choice_description'])
chipo_filtered.head()
```

### select only the products with quantity equals to 1
### select only the item_name and item_price columns
### sort the values from the most to less expensive

```python
chipo_one_prod = chipo_filtered[chipo_filtered['quantity'] == 1]
chipo_one_prod.head()
```

```python
chipo_one_prod.count()
```

```python
chipo_one_prod[['item_name', 'item_price']].sort_values(by='item_price', ascending = False).head(20)
```

#### select items with a price higher than $10

```python
chipo_one_prod[chipo_one_prod['item_price'] > 10]
```

### Step 6. Sort by the name of the item

```python
chipo_one_prod[['item_name', 'item_price']].sort_values(by='item_name').head(20)
```

### Step 7. What was the quantity of the most expensive item ordered?

```python
chipo[chipo['item_price'] == chipo['item_price'].max()]
```

### Step 8. How many times was a Veggie Salad Bowl ordered?

```python
chipo[chipo['item_name'] == 'Veggie Salad Bowl']['item_name'].count()
```

```python
len(chipo[chipo['item_name'] == 'Veggie Salad Bowl'])
```

### Step 9. How many times did someone order more than one Canned Soda?

```python
chipo['item_name'].value_counts()
```

```python
chipo[chipo['item_name'] == 'Canned Soda'] 
```

```python
len(chipo[(chipo['item_name'] == 'Canned Soda') & (chipo['quantity'] > 1)])
```
